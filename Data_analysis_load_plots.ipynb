```python
##https://stackoverflow.com/questions/20011122/fitting-a-normal-distribution-to-1d-data
import numpy as np
from scipy.stats import norm
import numpy as n
import matplotlib.pyplot as plt
import seaborn as sns
N, M = 10, 7
fig, ax = plt.subplots(1,figsize=(N, M))
# This is  the colormap I'd like to use.
cm = plt.cm.get_cmap('twilight_shifted')
#dm = plt.cm.get_cmap('cubehelix')

# Plot histogram.
n, bins, patches = plt.hist(df2, 25,
                           density=True,
                            color='green'
                           ,stacked = True
                           )
bin_centers = 0.5 * (bins[:-1] + bins[1:])
# scale values to interval [0,1]
col = bin_centers - min(bin_centers)
col /= max(col)
for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c))

#df.plot.kde(ax=ax, legend=False)
# Plot the PDF.
mu, std = norm.fit(df2)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=1,linestyle = 'dashdot')
#title = "Fit results: mu = %.2f,  std ,= %.2f" % (mu, std)
#plt.title(title)

# Setting the background color of the plot
# using set_facecolor() method

ax.set_facecolor("beige")

# Hide grid lines
ax.grid(False)

#/* Lighten borders */
ax.spines['bottom'].set_color('0.1')
ax.spines['top'].set_color('0.1')
ax.spines['right'].set_color('0.1')
ax.spines['left'].set_color('0.1')



ax.set_ylabel("Probability Density",
              #fontname="SimSun",
              fontsize=28)
ax.set_xlabel("Load (MW)",
              #fontname="SimSun",
              fontsize=28)

# Set the font name for axis tick labels to be Comic Sans
#for tick in ax.get_xticklabels():
 #   tick.set_fontname("SimSun")
#for tick in ax.get_yticklabels():
 #   tick.set_fontname('SimSun')

plt.xticks(fontsize=16, rotation=0)
plt.yticks(fontsize=16, rotation=0)

plt.tight_layout()

plt.savefig("Normal distribution.png", format="png", dpi=300)
```
