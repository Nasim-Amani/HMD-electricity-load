```python

import pandas as pd

# Read and store content of an excel file
read_file = pd.read_excel ("0--1.1.98 till 29.12.98.xlsx")

# Write the dataframe object into csv file
read_file.to_csv ("98.csv",
                  index = None,
                  header=True)

# read csv file and convert into a dataframe object
df = pd.DataFrame(pd.read_csv("98.csv"))
#df.to_csv('98.csv',  index = None)

# show the dataframe
df
```

```python
# read first line
with open("98.csv") as myfile:
    headRow = next(myfile)  # Read the first line of the file and assign it to the variable headRow

# read column names
columns = [x.strip() for x in headRow.split(',')]  # Split the line at each comma and remove leading/trailing whitespaces to get column names

# process by pandas
df2 = pd.read_csv("98.csv", names=columns[1:], skiprows=1)  # Read the CSV file, using column names starting from the second element and skipping the first row

df2_stacked = df2.stack()  # Stack the columns of the DataFrame into a single column

df2_stacked = df2_stacked.to_frame()  # Convert the stacked Series into a DataFrame
df2_stacked
```

```python
df2_stacked.columns=['Load']  # Renames the column of the DataFrame 'df2_stacked' to 'Load'.
df3=df2_stacked.reset_index()  # Resets the index of the DataFrame 'df2_stacked' and assigns the resulting DataFrame to 'df3'.
df3.columns=['Date','Hour','Load']  # Renames the columns of the DataFrame 'df3' to 'Date', 'Hour', and 'Load'.

df3
```

```python
df3.dtypes
```

split date column

```python
# Create a sample DataFrame with a column 'date'
data =df3['Date']
#df = pd.DataFrame(data)

# Split the 'date' column into separate year, month, and day columns
df3[['year', 'month', 'day']] = df3['Date'].str.split('/', expand=True)

# Print the updated DataFrame
print(df3)
```

```python
df3[['year', 'month', 'day']] = df3[['year', 'month', 'day']].astype(int)
df3.dtypes
```

```python

# Iterate over the rows of the DataFrame and convert each row to a Gregorian date
miladi_dates = []
for index, row in df3.iterrows():
    jalali_date = JalaliDate(row['year'], row['month'], row['day'])
    miladi_date = jalali_date.to_gregorian()
    miladi_dates.append(miladi_date)

# Create a new column 'miladi' in the DataFrame to store the converted dates
df3['miladi'] = miladi_dates

# Print the updated DataFrame
print(df3)
```

```python
# Extract the numeric part from the 'Hour' column
df3['Hour'] = df3['Hour'].str.extract('(\d+)')
df3
```

```python
columns_to_drop = ['Date', 'year', 	'month' ,	'day' 	]
df4 = df3.drop(columns=columns_to_drop)
df4
```

```python

# Convert the extracted numeric part to integers
df4['Hour'] = df4['Hour'].astype(int)

# Subtract 1 from the extracted numeric part
df4['Hour'] = df4['Hour'] - 1

# Add a colon (':') and append '00' to represent the minutes
df4['Hour'] = df4['Hour'].astype(str) + ':00'

# Print the updated DataFrame
print(df4)
```

```python
df4.columns=['Hour','Load','Date(Miladi)']  # Renames the columns of the DataFrame 'df3' to 'Date', 'Hour', and 'Load'.
```

```python
# Extract the last column
last_column = df4.iloc[:, -1]

# Remove the last column from the DataFrame
df4 = df4.drop(df4.columns[-1], axis=1)

# Insert the extracted column at the first position
df4.insert(0, 'Date(Miladi)', last_column)

# Print the updated DataFrame
print(df4)
```

```python
df4.dtypes
```

```python
# Assuming df4 is a DataFrame containing the columns 'Date(Miladi)', 'Hour', and 'Load'
df4["Date(Miladi)"] = pd.to_datetime(df4["Date(Miladi)"])
df4["Hour"] = pd.to_datetime(df4["Hour"])

# Combine the date and time columns into a single DateTime column
df4["DateTime"] = df4["Date(Miladi)"] + pd.to_timedelta(df4["Hour"].dt.hour, unit='H')
df4
```

```python
columns_to_drop = ['Date(Miladi)', 'Hour'	]
df5 = df4.drop(columns=columns_to_drop)
df5
```

```python
# Extract the last column
last_column = df5.iloc[:, -1]

# Remove the last column from the DataFrame
df6 = df5.drop(df5.columns[-1], axis=1)

# Insert the extracted column at the first position
df6.insert(0, 'DateTime', last_column)

# Print the updated DataFrame
print(df6)
```
