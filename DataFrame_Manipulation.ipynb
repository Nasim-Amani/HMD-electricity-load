# 98

```python

import pandas as pd

# Read and store content of an excel file
read_file = pd.read_excel ("0--1.1.98 till 29.12.98.xlsx")

# Write the dataframe object into csv file
read_file.to_csv ("98.csv",
                  index = None,
                  header=True)

# read csv file and convert into a dataframe object
df = pd.DataFrame(pd.read_csv("98.csv"))
#df.to_csv('98.csv',  index = None)

# show the dataframe
df
```

```python
# read first line
with open("98.csv") as myfile:
    headRow = next(myfile)  # Read the first line of the file and assign it to the variable headRow

# read column names
columns = [x.strip() for x in headRow.split(',')]  # Split the line at each comma and remove leading/trailing whitespaces to get column names

# process by pandas
df2 = pd.read_csv("98.csv", names=columns[1:], skiprows=1)  # Read the CSV file, using column names starting from the second element and skipping the first row

df2_stacked = df2.stack()  # Stack the columns of the DataFrame into a single column

df2_stacked = df2_stacked.to_frame()  # Convert the stacked Series into a DataFrame
df2_stacked
```

```python
df2_stacked.columns=['Load']  # Renames the column of the DataFrame 'df2_stacked' to 'Load'.
df3=df2_stacked.reset_index()  # Resets the index of the DataFrame 'df2_stacked' and assigns the resulting DataFrame to 'df3'.
df3.columns=['Date','Hour','Load']  # Renames the columns of the DataFrame 'df3' to 'Date', 'Hour', and 'Load'.
df3["DateTime"] = (df3["Date"] +" "+ df3["Hour"])  # Creates a new column 'DateTime' in 'df3' by combining the 'Date' and 'Hour' columns.
df3 = df3[["DateTime","Load"]]  # Reassigns 'df3' to a new DataFrame containing only the 'DateTime' and 'Load' columns.
df3.to_csv('98.csv',  index = None)
df3
```
